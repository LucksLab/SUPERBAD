```python
import tkinter as tk
from tkinter import ttk
from tkinter.messagebox import *
import xlrd

#Creates and shows the home window
root = tk.Tk()
root.geometry('500x300')
root.resizable(False, False)
root.title('Home')

#Creates the window for choosing the plate size
sizeWin = tk.Tk()
sizeWin.withdraw()   #Hides the size select window
sizeWin.geometry('500x300')
sizeWin.resizable(False, False)
sizeWin.title('Size Select')

#Creates the window for choosing duration of scan
durationWin = tk.Tk()
durationWin.withdraw()   #Hides the duration window
durationWin.geometry('500x300')
durationWin.resizable(False, False)
durationWin.title('Duration Select')

#Sets the window for choosing interval of each scan
intervalWin = tk.Tk()
intervalWin.withdraw()   #Hides the interval window
intervalWin.geometry('500x300')
intervalWin.resizable(False, False)
intervalWin.title('Interval Select')

#Creates variables for the 3 changeable parameters
selected_size = tk.StringVar()
duration_time = tk.DoubleVar()
interval_time = tk.DoubleVar()

#Function that returns you to the home window when the back button is selected
def back():
    #Hides the size select, duration, and interval windows
    sizeWin.withdraw()
    durationWin.withdraw()
    intervalWin.withdraw()
    #Shows the home window
    root.deiconify()

#Function that takes you to the plate size select window when the size select button is clicked
def size_select():
    #Destroys the widgets already associated with the size select window
    for widget in sizeWin.winfo_children():
        widget.destroy()
    
    #Hides the root window
    root.withdraw()
    #Shows the size select window
    sizeWin.deiconify()
    
    #Creates a label
    sizeLabel = ttk.Label(sizeWin, text="Select Size:")
    sizeLabel.pack(fill=tk.X, padx=5, pady=5)

    #Creates a combobox(drop down box)
    sizes = ttk.Combobox(sizeWin, textvariable=selected_size)

    #Sets the plate sizes you can choose from
    sizes['values'] = ['2x2', '4x4', '16x24']
    #Makes it so you cannot type in a plate size and must select from the given list
    sizes['state'] = 'readonly'

    #Organizes the widgets in blocks before placing them in the window
    sizes.pack(fill=tk.X, padx=5, pady=5)
    
    #Function that verifies the users selection and has them verify whether that is correct
    def confirm_size():
        #Gets the size selected
        selected_size = sizes.get()
        #Pop up box that ask you to verify whether the selected size is correct
        answer = askyesno(title='Confirmation', message='You selected the size ' 
                      + selected_size + ".\nContinue?")
        #Returns to the home screen if selected size is confirmed
        if answer is True:
            #Hides size select window
            sizeWin.withdraw()
            #Shows home window
            root.deiconify()

    #Creates the done button on the size select window
    sizeDoneButton = tk.Button(sizeWin, text='Done', fg='black', command=confirm_size)
    sizeDoneButton.config(height=3, width=20)
    sizeDoneButton.pack(fill=tk.X, padx=5, pady=5)
    
    #Creates the back button on the size select window
    sizeBackButton = tk.Button(sizeWin, text='Back', fg='black', command=back)
    sizeBackButton.config(height=3, width=20)
    sizeBackButton.pack(fill=tk.X, padx=5, pady=5)

#Function that takes you to the duration window
def duration():
    #Destroys the widgets already associated with the duration window
    for widget in durationWin.winfo_children():
        widget.destroy()
        
    #Hides home window
    root.withdraw()
    #Shows duration window
    durationWin.deiconify()
    
    #Creates a label
    durationLabel = tk.Label(durationWin, text="Enter duration in minutes:")
    durationLabel.pack(fill=tk.X, padx=5, pady=5)
    
    #Creates an input box for the user to input the duration of the scan in minutes
    duration_input = tk.Text(durationWin, height=5, width=5)
    duration_input.pack(fill=tk.X, padx=5, pady=5)
    
    #Function that verifies the users selection and has them verify whether that is correct
    def confirm_duration():
        #Gets the duration and formats it to a float value with 2 decimal places (ex: 5 mins becomes 5.00 mins)
        duration_time = "{:.2f}".format(float(duration_input.get(1.0, "end-1c")))
        #Pop up box that ask you to verify whether the inputed duration is correct
        duration_answer = askyesno(title='Confirmation', message='You selected a duration of '
                                   + str(duration_time) + ' minutes.\nContinue?')
        #Returns to the home screen if the inputted duration is confirmed
        if duration_answer is True:
            #Hides duration window
            durationWin.withdraw()
            #Shows home window
            root.deiconify()
            
    #Cretaes the done button on the duration window
    durationDoneButton = tk.Button(durationWin, text='Done', fg='black', command=confirm_duration)
    durationDoneButton.config(height=3, width=20)
    durationDoneButton.pack(fill=tk.X, padx=5, pady=5)
    
    #Creates the back button on the duration window
    durationBackButton = tk.Button(durationWin, text='Back', fg='black', command=back)
    durationBackButton.config(height=3, width=20)
    durationBackButton.pack(fill=tk.X, padx=5, pady=5)
    
#Function that takes you to the interval window
def intervals():
    #Destroys all widgets associated with the interval window
    for widget in intervalWin.winfo_children():
        widget.destroy()
        
    #Hides the home window
    root.withdraw()
    #Shows the interval window
    intervalWin.deiconify()
    
    #Creates a label
    intervalLabel = tk.Label(intervalWin, text="Enter interval length in minutes:")
    intervalLabel.pack(fill=tk.X, padx=5, pady=5)
    
    #Creates an input box for the user to input the interval scan time in minutes
    interval_input = tk.Text(intervalWin, height=5, width=5)
    interval_input.pack(fill=tk.X, padx=5, pady=5)
    
    #Function that verifies the users selection and has them verify whether that is correct
    def confirm_interval():
        #Gets the inputted interval time and converts it to a float value with 2 decimal places (ex: 5 mins becomes 5.00 mins)
        interval_time = "{:.2f}".format(float(interval_input.get(1.0, "end-1c")))
        #Pop up box that has the user confirm whether the selected interval time is correct
        interval_answer = askyesno(title='Confirmation', message='You selected interval lengths of '
                                   + str(interval_time) + " minutes.\nContinue?")
        #Returns to the home screen if the selected interval time is confirmed
        if interval_answer is True:
            #Hides interval window
            intervalWin.withdraw()
            #Shows home window
            root.deiconify()
        
    #Creates done button on interval window
    intervalDoneButton = tk.Button(intervalWin, text='Done', fg='black', command=confirm_interval)
    intervalDoneButton.config(height=3, width=20)
    intervalDoneButton.pack(fill=tk.X, padx=5, pady=5)
    
    #Creates back button on interval window
    intervalBackButton = tk.Button(intervalWin, text='Back', fg='black', command=back)
    intervalBackButton.config(height=3, width=20)
    intervalBackButton.pack(fill=tk.X, padx=5, pady=5)
    
    
#Creates the button that takes you to the plate size select window on the home window
select_size_button = tk.Button(root, text='Size Select', fg='black', command=size_select)
select_size_button.config(height=3, width=20)
select_size_button.pack(fill=tk.X, padx=5, pady=5)

#Creates the button that takes you to the duration of scan input window on the home window
duration_button = tk.Button(root, text='Duration', fg='black', command=duration)
duration_button.config(height=3, width=20)
duration_button.pack(fill=tk.X, padx=5, pady=5)

#Creates the button that takes you to the interval of scans input window on the home window
interval_button = tk.Button(root, text='Interval', fg='black', command=intervals)
interval_button.config(height=3, width=20)
interval_button.pack(fill=tk.X, padx=5, pady=5)

#Creates the exit button on the home window
#Exit button terminates the GUI
exit_button = tk.Button(root, text='Exit', command=root.quit)
exit_button.config(height=3, width=20)
exit_button.pack(fill=tk.X, padx=5, pady=5)

#Keeps the GUI running until terminated
root.mainloop()
#Terminates the GUI when the exit button is clicked
exit(0)
```
